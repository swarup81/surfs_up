```python
# Dependencies
import numpy as np
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
June_results = session.query(Measurement.date, Measurement.prcp).\
    filter(extract('month', Measurement.date)==6).all()
print(June_results)
```

```python
#Convert the June temperatures to a list.
june_temps_list = [temp.prcp for temp in June_results]
june_temps_list
```

```python
# Create a DataFrame from the list of temperatures for the month of June. 
june_df = pd.DataFrame(june_temps_list,columns=[ "June prcp"])
print(june_df)
```

```python
#Calculate and print out the summary statistics for the June temperature DataFrame.
june_df.describe()
```

```python
# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
Dec_results = session.query(Measurement.date, Measurement.prcp).\
    filter(extract('month', Measurement.date)==12).all()
print(Dec_results)
```

```python
 #Convert the December temperatures to a list.
dec_temps_list = [temp.prcp for temp in Dec_results]
dec_temps_list
```

```python
#  Create a DataFrame from the list of temperatures for the month of December. 
dec_df = pd.DataFrame(dec_temps_list, columns=[ "Dec prcp"])
print(dec_df)
```

```python
#Calculate and print out the summary statistics for the dec temperature DataFrame.
dec_df.describe()
```

```python
june_stats = june_df.describe()
dec_stats = dec_df.describe()
```

```python
# merge two precipitation statistics.
summary_df = june_stats.merge(dec_stats, left_index=True, right_index=True)
summary_df
```

```python
#list out the stations active for the month of june.
june_station = session.query(Measurement.tobs, Measurement.station).\
    filter(extract('month',Measurement.date)== 6).all()
print(june_station)
```

```python
#conver to dataframe
june_station_df = pd.DataFrame(june_station, columns=['june Temps', 'Station'])
june_station_df
```

```python
#list out the stations active for the month of dec.
dec_station = session.query(Measurement.tobs, Measurement.station).\
    filter(extract('month',Measurement.date)== 12).all()
print(dec_station)
```

```python
#convert to dataframe
dec_station_df = pd.DataFrame(dec_station, columns=['Dec Temps', 'Station'])
dec_station_df
```

```python
#merge two stations.
station = june_station_df.merge(dec_station_df, on="Station", how="inner").set_index("Station")
station
```
